```python
import pandas as pd
```

```python
Hero = "Resources/HeroesOfPymoli.csv"
```

```python
Hero_df = pd.read_csv(Hero, encoding="ISO-8859-1")
Hero_df.head()
```

```python
Totalplayervalue=Hero_df["SN"].nunique()
Total=pd.DataFrame({"Total Player":pd.Series(Totalplayervalue)})
Total
```

```python

#find number of unique items
uniqueitem = Hero_df["Item ID"].nunique()
#find average prices
averageprice=Hero_df["Price"].mean()
#find number of purchase
buy= Hero_df["Purchase ID"].count()
#find total revenue
totalrevenue= Hero_df["Price"].sum()


Purchase1=pd.DataFrame({"Number of Unique Items":pd.Series(uniqueitem),
                        "Average Price":pd.Series(averageprice),
                        "Number of Purchases":pd.Series(buy),
                        "Total Revenue":pd.Series(totalrevenue)})

#Display float with 2 decimal places
pd.options.display.float_format = '{:.2f}'.format
Purchase1





```

```python

#Count the number of players by gender and Redfine Columns
count = Hero_df.groupby('Gender')['SN'].nunique()

#Calculate the % of players by gender and Redfine Columns
percent=(Hero_df.groupby('Gender')['SN'].nunique())/(Hero_df.groupby('Gender')['SN'].nunique()).sum()

Demographics=pd.DataFrame({"Total Count":pd.Series(count),
                           "Percentage of Players":pd.Series(percent)})
Demographics1=Demographics.sort_values(by='Percentage of Players',ascending=False)
#Display % with 2 decimal 
pd.options.display.float_format = '{:.2%}'.format
Demographics1
```

```python
#calculate the Purchase Count
purchase_count = Hero_df.groupby('Gender')['Item Name'].count()

#Calucate the Average Purchase price
Price_by_Gender = Hero_df.groupby('Gender')['Price'].mean()


#Calucate the Total Purchase value and redfine columns
Purchase_by_Gender = Hero_df.groupby('Gender')['Price'].sum()

#Calucate the Avg Total Purchase per person 
Purchase_per_person=Purchase_by_Gender/count

Purchasing_Analysis=pd.DataFrame({"Purchase Count":pd.Series(purchase_count),
                                   "Average Purchase Price":pd.Series(Price_by_Gender),
                                   "Total Purchase Value":pd.Series(Purchase_by_Gender),
                                  "Avg Total Purchase per Person":pd.Series(Purchase_per_person)})
Purchasing_Analysis_sorted=Purchasing_Analysis.sort_values(by="Gender")
#Display float with 2 decimal places
pd.options.display.float_format = '{:.2f}'.format
Purchasing_Analysis_sorted
```

```python

#create a new table that drops duplicate SNs
newtable = Hero_df.drop_duplicates('SN')
#Create Age Bins
labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
bins = [0, 10, 15, 20, 25,30, 35, 40,50]
#Group into bins and puts labels and then count the values in each bin
age_count=pd.cut(newtable['Age'],bins=bins,labels=labels,right=False).value_counts()

# calculate the percent of each group
percent_age=age_count/Totalplayervalue

#create datatable
Age_Demographics=pd.DataFrame({"Total Count":pd.Series(age_count),
                                "Percent of Players":pd.Series(percent_age)})
#Display % with 2 decimal 
pd.options.display.float_format = '{:.2%}'.format
Age_Demographics.sort_index()
```

```python
#calculate Purchase count by age group
new_age_count=pd.cut(Hero_df['Age'],bins=bins,labels=labels,right=False).value_counts()

#create age bins 
age_bins=pd.cut(Hero_df['Age'],bins=bins,labels=labels,right=False)

#creat age bins and redefine columns
age_bins=pd.cut(Hero_df['Age'],bins=bins,labels=labels,right=False)
age_bins=age_bins.to_frame().reset_index()
age_bins.columns=['Purchase ID','Age Group']

#add age bins to Hero_df
temptable=pd.merge(Hero_df,age_bins,on='Purchase ID',)
# find total purchase by age group & format to 2 decimal places
total_purchase_age = temptable.groupby('Age Group')['Price'].sum()
total_purchase_age_format=total_purchase_age.map('${:,.2f}'.format)


# find average purchase by age group & format to 2 decimal places
total_purchase_mean=temptable.groupby('Age Group')['Price'].mean()
total_purchase_mean_format=total_purchase_mean.map('${:,.2f}'.format)
# find average purchase per person by age group 
total_average_per_person=total_purchase_age/age_count
total_average_per_person_format=total_average_per_person.map('${:,.2f}'.format)
#create summary table
purchasing_analysis_age=pd.DataFrame({"Purchase Count":pd.Series(new_age_count),
                                   "Average Purchase Price":pd.Series(total_purchase_mean_format),
                                    "Total Purchase ":pd.Series(total_purchase_age_format),
                                      "Avg Total Purchase per Person":pd.Series(total_average_per_person_format)})
purchasing_analysis_age.index.name="Age Ranges"
purchasing_analysis_age
```

```python

#calucalte Purchase Count
frequency=Hero_df["SN"].value_counts()

#Calculate Total Purchase
Total_purchase=Hero_df.groupby('SN')['Price'].sum()
Total_purchase_format=Total_purchase.map('${:,.2f}'.format)
#Calculate Average Purchase
Average_purchase_SN=Total_purchase/frequency
Average_purchase_SN_format=Average_purchase_SN.map('${:,.2f}'.format)
#create summary table
Top_Spender=pd.DataFrame({"Purchase Count":pd.Series(frequency),
                                   "Average Purchase Price":pd.Series(Average_purchase_SN_format),
                                    "Total Purchase Value ":pd.Series(Total_purchase_format),})
#Sort from Largest
Top_Spender_sorted=Top_Spender.sort_values(by="Purchase Count",ascending=False)
Top_Spender_sorted.index.name="SN"

Top_Spender_sorted.head()
```

```python

#create new table with Item ID, Item Name and Price
newHero_df = Hero_df.loc[:, ["Item ID", "Item Name","Price"]]

#afind total purchase value
Group_by=newHero_df.groupby(["Item ID", "Item Name"]).sum()

#count purchase count
Group_by["Purchase Count"]=pd.DataFrame(newHero_df.groupby(["Item ID", "Item Name"]).count())
#calculate price
Group_by["Item Price"]=Group_by['Price']/Group_by["Purchase Count"]

#rename price to Total Purchase Value
Most_Popular_df=Group_by.rename(columns={"Price":"Total Purchase Value"})
#Create Summary Table with column names
Most_Popular_df1=Most_Popular_df[["Purchase Count","Item Price","Total Purchase Value"]]
Most_Popular_df1["Item Price"]=Most_Popular_df1["Item Price"].map('${:,.2f}'.format)
Most_Popular_df1["Total Purchase Value"]=Most_Popular_df1["Total Purchase Value"].map('${:,.2f}'.format)
#Sort by Purchase Count
Most_Popular_df1_Sorted=Most_Popular_df1.sort_values(by="Purchase Count",ascending=False)
Most_Popular_df1_Sorted
```
